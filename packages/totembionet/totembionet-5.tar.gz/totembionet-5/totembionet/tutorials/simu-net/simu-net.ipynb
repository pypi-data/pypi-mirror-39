# Simu-Net

First we import both the discrete-model and simu-net library.
The discrete-model library is here to parse a model generated by SMBionet.

```python
import discrete_model
import simu_net
```

We parse all the models generated by SMBionet and we only keep the first one for this example.

```python
models = discrete_model.parse_smbionet_output_file('../resources/mucusOperonV2.out')
model, *_ = models
print(model)
```

Now we setup our simulation, we can configure the number of steps, the seed and the initial state.

```python
simulation = simu_net.Simulation(model)
simulation.steps = 10
simulation.random.seed(0xffff)
simulation.initial_state = {'operon': 1, 'mucuB': 0}
```

Now we are ready to run the simulation.

```python
result = simulation.run()
result.plot_evolution()
```
