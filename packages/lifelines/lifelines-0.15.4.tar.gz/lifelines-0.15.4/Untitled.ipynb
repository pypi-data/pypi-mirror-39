```python
# loglikelihood:

from sympy import *
gamma = functions.special.gamma_functions.gamma
lower_gamma = functions.special.gamma_functions.lowergamma

a, b, x, d = symbols(', b, x, d')

def pdf(x):
    return b**a / gamma(a) * x ** (a-1) * exp(-b * x)


def cdf(x):
    return lower_gamma(a, b*x) / gamma(a)

def sf(x):
	return 1 - cdf(x)

hazard = -diff(sf(x), x)/sf(x)
c_hazard = -log(sf(x))
```

```python
ll = d * log(hazard) - c_hazard
```

```python
diff(ll, a, a)
```
